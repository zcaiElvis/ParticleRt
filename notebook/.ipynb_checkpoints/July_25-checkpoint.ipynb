```r
y_pred <- one_day_pred(rt = ca_result$pred.mean.x, y= ca$y, shape = nz_shape, scale= nz_scale)
```

```r
ggplot(data=data.frame(y=ca$y, y_pred = y_pred, idx = ca$idx), aes(x = idx, y=y ))+
geom_line(aes(y=y), color = "blue")+
geom_point(aes(y=y_pred), color = "red")+
theme_bw()
```

# New Zealand Cases

```r
alldata = read.csv('data/WHO-COVID-19-global-data.csv')
idcountry = which(alldata$Country == 'New Zealand')
Iday = alldata$New_cases[idcountry]
Iday <- rollmean(Iday, k=7, fill=NA)
Iday <- na.omit(Iday)
covid_nz <- data.frame(y = round(Iday), idx = 1:length(Iday))
```

```r
plot(covid_nz$y)
```

```r
nz_filter <- run_pfilter(rprocess_config = tran_lognormal, dmeasure_config = meas_pois_interval, data = covid_nz,
                       Np = 500, x_init = 1, shape = 2.3669, scale = 2.7463)
```
