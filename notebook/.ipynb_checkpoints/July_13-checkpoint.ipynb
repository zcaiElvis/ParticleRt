# Pomp example - Normal transition + Poisson measurement

### Try switching `tran_normal_unconst` to `tran_normal_const`. 
1. Truncated normal is really bad
2. Sometimes unconstrainted process samples values below 0. Data generated for this example starts from 50 to make sure that this doesn't happen
3. Data generation process uses truncated normal as well
4. $eff \in [299,300]$

```r
filter_trnorm_pois <- run_pfilter(rprocess_config = tran_normal_unconst, dmeasure_config = meas_pois,
                                data = sim_trnorm_pois, Np = 300, sigma1 = 1, sigma2 = 1, x_init = 100)
```

```r
plot(filter_trnorm_pois)
```

```r
filtered_trnorm_pois_x <- as.data.frame(filter_trnorm_pois)$pred.mean.x
```

```r
plot(filtered_trnorm_pois_x)
plot(sim_trnorm_pois_x)
```

```r
plot(filtered_trnorm_pois_x, sim_trnorm_pois_x)
```

## Simulate dataset
1. Data generation: a good dataset is extremely hard to simulate: see counter below for the number of simulations

```r
sim_lnorm_pois_rw <- gen_lnormal_pois_rw(size = 1000, sd_log = 0.02, init_x = 1, g_shape = 2, g_scale = 2)

counter <- 0
while(max(sim_lnorm_pois_rw$y) < 2000 | max(sim_lnorm_pois_rw$y)> 100000000){
    sim_lnorm_pois_rw <- gen_lnormal_pois_rw(size = 1000, sd_log = 0.02, init_x = 1, g_shape = 2, g_scale = 2)
    counter <- counter + 1
}
counter
```

```r
plot(sim_lnorm_pois_rw$y)
plot(sim_lnorm_pois_rw$x)
```

```r
sim_lnorm_pois_x <- sim_lnorm_pois_rw$x
sim_lnorm_pois_rw <- select(sim_lnorm_pois_rw, all_of(c("idx", "y")))
```

```r
sim_lnorm_pois_rw$y <- log(sim_lnorm_pois_rw$y)
```

```r
pfilter_lnorm_pois <- run_pfilter(rprocess_config = tran_lognormal, dmeasure_config = meas_pois_lnorm,
                                data = sim_lnorm_pois_rw, Np = 500, 
                                sigma1 = 1, sigma2 = 1, sdlog = 0.02, x_init = 1, shape = 2, scale = 2)
```

## Result analysis - Plot below
1. We know whole data generation process
2. $eff \in [499, 500]$
3. Conditional likelihood $Pr(Y_j | Y_1^{j-1})$ goes down rapidly after 400 -> estimate after that are not reliable
4. Possible solutions: inflate the variance of poisson, or inflate the variance of lognormal
    - Inflating variance of poisson affect $R_t$ as well
    - Lognormal: variance changes as mean of lognormal changes as well. Could add another inflation parameter, much like the i=0.1 in epifilter

```r
plot(pfilter_lnorm_pois)
```

```r
filtered_lnorm_pois_x <- as.data.frame(pfilter_lnorm_pois)$pred.mean.x
```

```r
plot(filtered_lnorm_pois_x, sim_lnorm_pois_x)
```
